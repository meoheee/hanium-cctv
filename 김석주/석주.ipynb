```python
import cv2
import matplotlib.pyplot as plt

img_basic = cv2.imread('pi.jpg', cv2.IMREAD_COLOR)
plt.imshow(cv2.cvtColor(img_basic, cv2.COLOR_BGR2RGB))
plt.show()

img_basic = cv2.cvtColor(img_basic, cv2.COLOR_BGR2GRAY)
plt.imshow(cv2.cvtColor(img_basic, cv2.COLOR_GRAY2RGB))
plt.show()
```

```python
import cv2

cam = cv2.VideoCapture(0)
a = cv2.VideoWriter('vid.mp4',cv2.VideoWriter_fourcc(*'XVID'), 30.0, (1280, 720))
while 1 :
    ret,img = cam.read()
    if ret == False:
        break
    grey = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    cv2.imshow('video', img)
    cv2.imshow('grey', grey)
    a.write(img)
    a.write(grey)
    if cv2.waitKey(1)&0xFF == 27:
        break
cam.release()
a.release()
cv2.destroyAllWindows()
```
